```python
from datasets import load_dataset
```

```python
dataset_v2 = load_dataset("presencesw/hash_20_256_v2", streaming=True, split='train')
# dataset_v2 = dataset_v2.skip(4*3)
```

```python
check_data = dataset_v2.skip(int(10_000*8/2))
check_data = check_data.take(int(10_000/2))
```

```python
check_data = check_data.shuffle(42,buffer_size=10_000)
check_data = check_data.skip(int((1_200-12*14)/2))
```

```python
a1 = list(check_data.take(int(12*14/2)))
new_check_data = check_data.skip(int(12*14/2))
a2 = list(new_check_data.take(int(12*14/2)))
```

```python
print(a1)
print(a2)
```

```python
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained("google/mt5-large")
```

```python
for i in a1:
    print(i['vi'])
    print(i['en'])
```

```python
for i in a2:
    print(i['vi'])
```

```python
print(a1[0]['en'])
print(a1[0]['vi'])
```

```python
for i in a1:
    if len(tokenizer.tokenize(i['en'], return_tensors="pt")) > 256:
        print(True)
    if len(tokenizer.tokenize(i['vi'], return_tensors="pt")) > 256:
        print(True)
    if len(tokenizer.tokenize(i['vi'], return_tensors="pt")) - len(tokenizer.tokenize(i['en'], return_tensors="pt")) >= 50 or len(tokenizer.tokenize(i['vi'], return_tensors="pt")) - len(tokenizer.tokenize(i['en'], return_tensors="pt")) <= -50:
        print(i['en'])
        print(len(tokenizer.tokenize(i['en'], return_tensors="pt")))
        print(i['vi'])
        print(len(tokenizer.tokenize(i['vi'], return_tensors="pt")), "\n")
```

```python
for i in a2:
    if len(tokenizer.tokenize(i['en'], return_tensors="pt")) > 256:
        print(True)
    if len(tokenizer.tokenize(i['vi'], return_tensors="pt")) > 256:
        print(True)
    if len(tokenizer.tokenize(i['vi'], return_tensors="pt")) - len(tokenizer.tokenize(i['en'], return_tensors="pt")) >= 50 or len(tokenizer.tokenize(i['vi'], return_tensors="pt")) - len(tokenizer.tokenize(i['en'], return_tensors="pt")) <= -50:
        print(i['en'])
        print(len(tokenizer.tokenize(i['en'], return_tensors="pt")))
        print(i['vi'])
        print(len(tokenizer.tokenize(i['vi'], return_tensors="pt")), "\n")
```
